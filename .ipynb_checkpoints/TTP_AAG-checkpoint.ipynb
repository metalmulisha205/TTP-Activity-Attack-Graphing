```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from efficient_apriori import apriori
from pyattck import Attck
```

```python
df = pd.read_csv("Categorized_Adversary_TTPs.csv")
df.head()
```

```python
# to use the apriori we need to generate a list of lists
columns = df.columns
aprList = []
for row in df.values:
    aprList.append((row[11].strip('][').split(', ') + row[13].strip('][').split(', ')))
```

```python
for i in range(0,4):
    print(aprList[i])
```

```python
#note: takes a long time to run with just 5k observations. Might need to find another algorithm or make my own implementation if this is going to be successful on a larger dataset.
itemsets, rules = apriori(aprList, min_support=0.1, min_confidence=0.80)
```

```python
# sorting the rules by confidence for now
confs = np.array([rule.confidence for rule in rules])
rules = np.array(rules)
inds = confs.argsort()[::-1]
rules = rules[inds]
```

```python
# each rule has a confidence, lift, support, lhs, rhs, conviction, rule power factor (RPF), 
# lots of rules with multiple lhs and/or multiple rhs
# how can we visualize this and what can we learn from this
# TODO: function that can calculate the support, lift, and confidence for a given rule. Let it take a tuple of 2 tuples for a lhs and rhs
for rule in rules:
    if (len(rule.lhs) == 1 and rule.lhs[0][1] != "T"):
        print(str(rule) + "\n")
        #pass
```

```python
# filter for only 1-1 mappings 
rules11 = list(filter(lambda rule: len(rule.lhs) == 1 and len(rule.rhs) == 1 and rule.confidence < 1.00, rules))
urules11 = [[rule.lhs, rule.rhs] for rule in rules11]
urules11 = list(set(np.array(urules11).flatten()))
for rule in rules11:
    if (len(rule.rhs) > 0):
        print(str(rule) + "\n")
```

```python
import requests

def get_technique_name(technique_id):
  attack_api_url = 'https://api.mitre.org/api/attack/v3/techniques/{}'.format(technique_id)
  response = requests.get(attack_api_url)
  if response.status_code == 200:
    technique = response.json()
    return technique['name']
  else:
    return None

technique_id = 'T1041'
technique_name = get_technique_name(technique_id)

print(technique_name)
```
